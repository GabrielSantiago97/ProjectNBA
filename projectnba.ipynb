```python
import io
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(io.BytesIO(uploaded['2017-18_NBA_salary.csv']))
```

##ANALISAR DADOS

```python
#Visualizar dados
df.head()
```

```python
#Colunas, qtde de linhas e dtype
df.info()
```

```python
#Excluir colunas que não serão usadas nesse projeto
df = df.drop(columns=['VORP', 'BPM', 'DBPM', 'OBPM', 'WS/48', 'USG%', 'DWS', 'OWS', 'PER', 'TS%', 'WS'])
```

```python
#Player com maior salário... TOP 5
playersal = df[["Player", "Salary"]].groupby("Player").sum()
playersal.sort_values(by="Salary", ascending=False).head()
```

```python
#Salario dos jogadores que foram 1º draft
draft = df.loc[(df['NBA_DraftNumber'] == 1)]
draft
```

```python
#Ranking top 5 jogadores com maiores salários 1º pick draft
draftsal = draft[["Player", "Salary"]].groupby("Player").sum()
draftsal.sort_values(by="Salary", ascending=False).head()
```

```python
#Verificar quantos países estão na base
countryclear = df.drop_duplicates(subset=['NBA_Country'])
country = countryclear['NBA_Country']
country.shape[0]
```

```python
#Verificar se existe algum jogador draftado do brasil
brasil = countryclear.loc[(countryclear['NBA_Country'] == 'Brazil')]
brasil
```

```python
#Qual jogador teve maior número de assistencias
totalpass = df[["Player", "AST%"]].groupby("Player").sum()
totalpass.sort_values(by="AST%", ascending=False).head(1)
```

```python
#Jogador mais experiênte
agexp = df[["Player", "Age"]].groupby("Player").mean()
agexp.sort_values(by="Age", ascending=False).head(1)
```

```python
#Criar colunas com sobrenomes dos jogadores
df['SecondName'] = df['Player'].str.split().str[1]
df.head()
```

```python
#Cria e-mail ficticio com sobrenome dos jogadores e domínio @gmail.com
def criar_email(nome):
    return f'{nome}@gmail.com'

df['Email'] = df['SecondName'].apply(criar_email)
df.head()
```
